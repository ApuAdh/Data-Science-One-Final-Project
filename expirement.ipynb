```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#Getting the names of all pokemon
names = []

payload = requests.get("https://pokemondb.net/pokedex/all")
soup = BeautifulSoup(payload.text, 'html.parser')
cells = soup.find_all('td', class_='cell-name')
for cell in cells:
    link = cell.find('a', class_='ent-name')
    small_text = cell.find('small', class_='text-muted')
    if small_text:
        names.append(small_text.text.strip())
    elif link:
        names.append(link.text.strip())
```

```python
# Getting the types of all pokemon
type1 = []
type2 = []

cells = soup.find_all('td', class_='cell-icon')
for cell in cells:
    links = cell.find_all('a', class_='type-icon')
    type1.append(links[0].text.strip())
    if len(links) > 1:
        type2.append(links[1].text.strip())
    else:
        type2.append(None)
```

```python
#Getting the following values of the pokemon
Pokedex_ID, BST, HP, Attack, Defence, Sp_Attack, Sp_Defence, Speed = [], [], [], [], [], [], [], []
cells = soup.find_all('td', class_='cell-num')
for idx, cell in enumerate(cells):
    value = cell.text.strip()
    if idx % 8 == 0:
        Pokedex_ID.append(int(value))
    elif idx % 8 == 1:
        BST.append(int(value))
    elif idx % 8 == 2:
        HP.append(int(value))
    elif idx % 8 == 3:
        Attack.append(int(value))
    elif idx % 8 == 4:
        Defence.append(int(value))
    elif idx % 8 == 5:
        Sp_Attack.append(int(value))
    elif idx % 8 == 6:
        Sp_Defence.append(int(value))
    elif idx % 8 == 7:
        Speed.append(int(value))
```

```python
dict = {
    'pokedex_id': Pokedex_ID,
    'name':names,
    'type1': type1,
    'type2': type2,
    "hp": HP,
    'attack': Attack,
    'defence': Defence,
    'special_attack': Sp_Attack,
    'special_defence': Sp_Defence,
    'speed': Speed
}
df = pd.DataFrame.from_dict(dict)
```

```python
display(df)
```

```python
def TypeSpecificity(type):
    filtered_df = df[(df["type1"] == type) | (df["type2"] == type)]
    filtered_df = filtered_df.assign(
        Preferred_Attack = filtered_df[['attack', 'special_attack']].max(axis=1)
    )
    return filtered_df
```

```python
type = TypeSpecificity("Steel")
name = []
for i in type["name"]:
    name.append(i)
print(name)
```

```python
preferred_attack = []
speed = []
for i in type["Preferred_Attack"]:
    preferred_attack.append(i)
for i in type["speed"]:
    speed.append(i)
for i in range(len(preferred_attack)):
    print("{x:"+str(preferred_attack[i])+",y:"+str(speed[i])+"},")
```
